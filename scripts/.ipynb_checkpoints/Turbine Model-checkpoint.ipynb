```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from smt.surrogate_models import KRG
from sklearn.model_selection import ParameterGrid
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import (RBF, Matern, RationalQuadratic,
                                              ExpSineSquared, DotProduct,
                                              ConstantKernel, WhiteKernel)
import glob
import os
import pygmo as pg
from smt.surrogate_models import KRG
import random
```

```python
class turbine_GP_reg(object):
    
    def __init__(self, file_path, y_col = 3, quadratic = False):

        all_files = glob.glob(os.path.join(file_path, "*.csv")) #make list of paths
        self.data_loaded = False
        self.quadratic = quadratic
        
        print("loading files: ")
        for file in all_files:
            file_name = os.path.splitext(os.path.basename(file))[0]  # Getting the file name without extension
            if "15" in file_name:
                dataframe = pd.read_csv(file)

                dataframe['Speed (rad/s)'] = dataframe['Speed (RPM)']*0.104719755
                dataframe['torque_adj (mNm)'] = dataframe['I (A)']*47.9953+16.03362
                dataframe['turbine power (w)'] = dataframe['Speed (rad/s)'] * dataframe['torque_adj (mNm)'] / 1000
                dataframe['fluid power (w)'] = dataframe['Flow Rate (GPM)'] / 15.89806024 * (dataframe['DP (psi)']-0.491)*6894.76/997
                dataframe['turbine eff'] = dataframe['turbine power (w)'] / dataframe['fluid power (w)']


                x = dataframe[['Speed (rad/s)','GV','Flow Rate (GPM)','DP (psi)','torque (mNm)','turbine eff']]

                if self.data_loaded:
                    self.data_df = pd.concat([self.data_df, x])
                    self.data_all_df = pd.concat([self.data_all_df, dataframe])
                else:
                    self.data_loaded=True
                    self.data_df = x
                    self.data_all_df = dataframe

                print(file_name)
                print(dataframe.shape)
            
        self.data_all_df.to_csv("combined_data.csv")
        self.data_df = self.data_df.sort_values('GV')
        self.data = np.array(np.array(self.data_df.values))
        print("\ntotal num of data pts: " +str(len(self.data))+'\n')
        
        self.X_train = []
        self.y_train = []
        self.X_test = []
        self.y_test = []
        
        self.model = None
        self.model_RBF_scale = [1, 3, 7]
        self.model_noise = 1e-5
        
        self.trainning_error = 0
        self.test_error = 0
        
        self.data_split = 1
        self.y_col = y_col
        self.len = len(self.data)
        self.limits = []
    
    def train_GP(self):
        
        if not self.quadratic:
            kernel= 1.0 * RBF(length_scale=self.model_RBF_scale, length_scale_bounds=(1e-3, 100000.0)) + WhiteKernel(noise_level=self.model_noise, noise_level_bounds=(1e-10, 1e+10))
            self.model = GaussianProcessRegressor(kernel=kernel)
            self.model.fit(self.X_train, self.y_train)
        else:
            
            self.model = KRG(theta0=self.model_RBF_scale,poly = 'quadratic', print_training=False,print_global=False)  
            self.model.set_training_values(self.X_train, self.y_train)
            self.model.train()

        return self.model
    
    def get_train_error(self):
        if not self.quadratic:
            return np.sum((self.model.predict(self.X_train)-self.y_train)**2)
        else:
            return np.sum((self.model.predict_values(self.X_train)-self.y_train)**2)
    
    def get_test_error(self):
        if not self.quadratic:
            return np.sum((self.model.predict(self.X_test)-self.y_test)**2)
        else:
            return np.sum((self.model.predict_values(self.X_test)-self.y_test)**2)
    
    def train_with_crossvalidation(self, split=7):
        
        self.data_split = split
        self.trainning_error = 0
        self.test_error = 0
        
        for i in range(self.data_split):
            start = i*self.len//self.data_split
            if i == self.data_split - 1:
                end = self.len
            else:
                end = (i+1)*self.len//self.data_split
                
            train, test = np.concatenate([self.data[0:start,:], self.data[end:,:]]), self.data[start:end,:]
            
            self.X_train = train[:,0:3]
            self.y_train = train[:,self.y_col:self.y_col+1]
            self.X_test = test[:,0:3]
            self.y_test = test[:,self.y_col:self.y_col+1]

            self.train_GP()
            
            self.trainning_error += self.get_train_error() * 1/self.data_split/len(self.X_train)
            self.test_error += self.get_test_error() * 1/self.data_split/len(self.X_test)
        
        return [self.trainning_error, self.test_error]
    
    def full_train_GP(self, model_RBF_scale, model_noise):
        self.model_RBF_scale = model_RBF_scale
        self.model_noise = model_noise
        
        self.X_train = self.data[:,0:3]
        self.y_train = self.data[:,self.y_col:self.y_col+1]
        self.train_GP()
        self.get_train_error()
        
        print("trainning error: " + str(self.trainning_error))
        
        return True
    
    def grid_search(self):
        
        RFB_scale_grid = [1,3,4,5,7,9]
        noise_grid = [1e-5]
        
        param_grid = {'param1': RFB_scale_grid, 'param2' : RFB_scale_grid, 'param3':RFB_scale_grid, 'param4':noise_grid}
        
        grid = ParameterGrid(param_grid)
        
        #np.random.shuffle(self.data)
        
        best_settings =[]
        best_test_error = 99999
        for params in grid:
            self.model_RBF_scale = [params['param1'], params['param2'], params['param3']]
            self.noise_level = params['param4']
            
            print("\nhyperparameters: ")
            print(self.model_RBF_scale)
            print(self.noise_level)
            
            self.train_with_crossvalidation()
            
            print("trainning error: " + str(self.trainning_error))
            print("testing error: " + str(self.test_error))
            
            if self.test_error < best_test_error:
                best_settings = self.model_RBF_scale + [self.noise_level]
                best_test_error = self.test_error
        
        self.full_train_GP(best_settings[:3], best_settings[3])
        
        print("\nbest hyperparameter settings: ")
        print(best_settings)
        print("\nbest testing error: ")
        print(best_test_error)
        print("final training error: " + str(self.trainning_error))
        
        return True
    
    def get_limits(self):

        GVs = self.data_df.GV.unique()
        for GV_angle in GVs:
            temp = [GV_angle]
            temp.append(self.data_df[(self.data_df.GV == GV_angle)]['Speed (rad/s)'].min())
            temp.append(self.data_df[(self.data_df.GV == GV_angle)]['Speed (rad/s)'].max())
            self.limits.append(temp)
        
        return self.limits
    
    def save_data(self):
        
        self.data_all_df.to_csv("combined_data.csv")
        
        return True
```

```python
class turbine_steady_state_model(object):
    
    def __init__(self, DP_model, torque_model, DP_model_GP, torque_model_GP, torque_constant= 45.159, torque_resistance=6.3329,\
                 speed_constant = 216.943, internal_resistance = 2.434):
        
        self.DP_model = DP_model
        self.torque_model = torque_model
        
        self.DP_model_GP = DP_model_GP
        self.torque_model_GP = torque_model_GP
        
        self.torque_constant = torque_constant #mNm/A
        self.torque_resistance = torque_resistance #mNm
        
        self.speed_constant = speed_constant * 0.104719755 #rad/s/V
        self.internal_resistance = internal_resistance
        
        self.G = 0
        self.w = 0
        self.q = 0
        self.I = 0
        
        self.Tm = 0
        self.DP = 0
        self.Te = 0
        self.V = 0
        
    def update_states(self,G=0, w=0, q=0, I=0):
        self.G = G
        self.w = w
        self.q = q 
        self.I = I
        
        self.update_DP()
        self.update_Tm()
        self.update_Te()
        self.update_V()
        
    def update_Te(self):
        self.Te = self.torque_constant*self.I + self.torque_resistance #mNm
    
    def update_V(self):
        self.V = self.w / self.speed_constant - self.internal_resistance * self.I  #V
    
    def update_Tm(self):
        mean, std = self.torque_model_GP.model.predict(np.array([[self.w,self.G,self.q]]), return_std=True)
        if std[0] < 8.0:
            self.Tm = mean[0][0]
        else:
            self.Tm = self.torque_model.model.predict_values(np.array([[self.w,self.G,self.q]]))[0][0]
        
    def update_DP(self):
        mean, std = self.DP_model_GP.model.predict(np.array([[self.w,self.G,self.q]]), return_std=True)
        if std[0] < 2.0:
            self.DP = mean[0][0]
        else:
            self.DP = self.DP_model.model.predict_values(np.array([[self.w,self.G,self.q]]))[0][0]
    
    def get_current_states(self):
        return [self.G, self.w, self.q, self.DP, self.Tm, self.I, self.V, self.Te]
```

```python
class turbine_dynamic_model(object):
    
    def __init__(self, Tw = 0.1 , H = 0.8):
        
        self.Tw = Tw
        self.H = H
        
        self.state_t = []
        self.current_state = None
        
    def simulate(self, init, t, G_t, I_t):
        
        self.current_state = init
        self.state_t = []
        self.state_t.append([t[0]] + init.get_current_states())
        self.state_t.append([t[0]] + init.get_current_states())
        
        for i in range(1,len(t)):
            dt = t[i] - t[i-1]
            w_next = self.intergrate_w(dt)
            q_next = self.intergrate_q(dt)
            
            self.current_state.update_states(G=G_t[i], w=w_next, q=q_next, I=I_t[i])
            self.state_t.append([t[i]] + self.current_state.get_current_states())
        
        self.state_t.pop(0)
            
    def intergrate_w(self, dt):
        return self.current_state.w + 1/self.H * (self.current_state.Tm - self.current_state.Te)*dt
    
    def intergrate_q(self, dt):
        return self.current_state.q - 1/self.Tw * (self.current_state.DP - self.state_t[-2][4])*dt
```

```python
class turbine_system_identification(object):
    
    def __init__(self, file_path, DP_model, torque_model, DP_model_GP, torque_model_GP, start = 20, objectives = ['Speed (rad/s)', 'Flow Rate (GPM)',\
                                                                                   'torque (mNm)','DP (psi)', 'V (V)']):
        
        all_files = glob.glob(os.path.join(file_path, "*.csv")) #make list of paths
        self.dynamic_data=[]
        self.filenames = [] 
        print("loading files: ")
        for file in all_files:
            file_name = os.path.splitext(os.path.basename(file))[0]  # Getting the file name without extension
            if 'GV' in file_name:
                dataframe = pd.read_csv(file)
                dataframe['Speed (rad/s)'] = dataframe['Speed (RPM)']*0.104719755
                x = dataframe[['Time (sec)', 'GV', 'I (A)', 'Speed (rad/s)', 'Flow Rate (GPM)', 'DP (psi)','torque (mNm)', 'V (V)' ]]
                self.dynamic_data.append(x)
                self.filenames.append(file_name)
                print(file_name)
                print(dataframe.shape)
        print()
            
        self.DP_model = DP_model
        self.torque_model = torque_model
        self.DP_model_GP = DP_model_GP
        self.torque_model_GP = torque_model_GP
        
        self.predicted_data=[]
        self.start = start
        self.objectives = objectives
        self.progress = 0
        
    def get_prediction(self, model_parameters):
        
        self.predicted_data = []
        
        dynamic_model = turbine_dynamic_model(Tw = model_parameters[0] , H = model_parameters[1])
        
        steady_state_model = turbine_steady_state_model(self.DP_model, self.torque_model,\
                                                        self.DP_model_GP, self.torque_model_GP,\
                                                        torque_constant = model_parameters[2],\
                                                        torque_resistance = model_parameters[3],\
                                                        speed_constant = model_parameters[4],\
                                                        internal_resistance = model_parameters[5])
        
        for i, runs in enumerate(self.dynamic_data):
            #print("running file: " + self.filenames[i])
            data = runs.values
            steady_state_model.update_states(G=data[0:self.start,1].mean(), w=data[0:self.start,3].mean(),\
                                             q=data[0:self.start,4].mean(), I=data[0:self.start,2].mean())
            
            t= data[self.start:,0]
            G_t = data[self.start:,1]
            I_t = data[self.start:,2]
            
            dynamic_model.simulate(steady_state_model, t, G_t, I_t)
            df = pd.DataFrame(dynamic_model.state_t, columns =['Time (sec)', 'GV', 'Speed (rad/s)', 'Flow Rate (GPM)',\
                                                               'DP (psi)','torque mech (mNm)', 'I (A)', 'V (V)',\
                                                               'torque (mNm)' ]) 
            self.predicted_data.append(df)
            
    def prediction_error(self, weight = 1):
        error = [0, 0, 0, 0, 0]
        for i, runs in enumerate(self.dynamic_data):
            #print("evaluating file: " + self.filenames[i])
            size = len(self.predicted_data[i][:]['Time (sec)'].values)
            for j, variable in enumerate(self.objectives):
                error[j] += np.sum((runs[self.start:][variable].values - self.predicted_data[i][:][variable].values)**2) / size
        return error
    
    def fitness(self,model_parameters):
        param = []
        for p in model_parameters:
            param.append(p)
        try:
            self.get_prediction(param)
            error = self.prediction_error()
        except:
            print(param)
            error = [100000, 100000, 100000, 100000, 100000]
            
        self.progress += 1
        if self.progress%32 == 0:
            print(self.progress//32)
        return error
    
    def get_bounds(self):
        return ([1,0.1,40,1,200,0.5],[20,10,50,30,300,10]) #problem.evaluate_model_parameters([2, 0.8, 46.4, 6.3329, 206, 55.312])
    
    def get_nobj(self):
        return len(self.objectives)
```

```python
def plot_surface(model,zlabel,range_x=[[300,650],[0,20]], grid_x = 20, grid_y = 20,flow=27.8, quadratic=False):
        
    if not quadratic:
        limits = [[0.0, 150, 460],[12, 455, 556], [15, 500, 556],[20.0, 537, 620]]
    else:
        limits = [[0.0, 190, 460], [20, 500, 620]]
    
    grid_limits = []
    d_angle = (range_x[1][1]-range_x[1][0])/grid_x
    for i in range(grid_x):
        temp = [range_x[1][0] + i*d_angle]
        j=1
        while temp[0] > limits[j][0]:
            j+=1
        interpolation_ratio = (temp[0] - limits[j-1][0]) / (limits[j][0] - limits[j-1][0])
        temp.append(limits[j-1][1] + (limits[j][1] - limits[j-1][1])*interpolation_ratio)
        temp.append(limits[j-1][2] + (limits[j][2] - limits[j-1][2])*interpolation_ratio)
        grid_limits.append(temp) 
    #print(grid_limits)
    grid_x_FF = np.zeros((grid_x,grid_y))
    grid_y_FF = np.zeros((grid_x,grid_y))
    for i in range(grid_x):
        grid_y_FF[i,:] = np.linspace(grid_limits[i][0],grid_limits[i][0],grid_x)
        grid_x_FF[i,:] = np.linspace(grid_limits[i][1],grid_limits[i][2],grid_x)

    
    Z_FF=np.zeros((grid_x,grid_y))
    for i in range(grid_x):
        for j in range(grid_y):
            if not quadratic:
                Z_FF[i][j]=model.model.predict(np.array([[grid_x_FF[i][j],grid_y_FF[i][j],flow]]))[0][0]
            else:
                Z_FF[i][j]=model.model.predict_values(np.array([[grid_x_FF[i][j],grid_y_FF[i][j],flow]]))[0][0]
            
    fig1 = plt.figure(figsize=(7,6))
    ax1 = axes3d.Axes3D(fig1)    
    ax1.plot_wireframe(grid_x_FF, grid_y_FF, Z_FF,cmap="YlGnBu_r")
    ax1.plot_surface(grid_x_FF, grid_y_FF, Z_FF,alpha=0.5, antialiased=True,cmap="YlGnBu_r")
    #ax1.scatter3D(x_FF,y_FF,z_FF, c='b')
    ax1.set_xlabel('Angular Speed (rad/s)', fontsize= 14, labelpad=10)
    ax1.set_ylabel('GV Angle (Deg)', fontsize= 14, labelpad=10)
    ax1.set_zlabel(zlabel, fontsize= 14,  labelpad=5.5, rotation=300)
    #ax1.tick_params(which = 'both', pad=1, labelsize=12)
    ax1.view_init(45, -70)    
    
    plt.autoscale()
    plt.savefig(zlabel + '.png' , bbox_inches='tight', dpi=200)
```

```python
class turbine_system_identification_SO(object):
    
    def __init__(self, file_path, DP_model, torque_model, DP_model_GP, torque_model_GP, weight, start = 20, objectives = ['Speed (rad/s)', 'Flow Rate (GPM)',\
                                                                                   'torque (mNm)','DP (psi)', 'V (V)']):
        
        all_files = glob.glob(os.path.join(file_path, "*.csv")) #make list of paths
        self.dynamic_data=[]
        self.filenames = [] 
        print("loading files: ")
        for file in all_files:
            file_name = os.path.splitext(os.path.basename(file))[0]  # Getting the file name without extension
            if 'GV' in file_name:
                dataframe = pd.read_csv(file)
                dataframe['Speed (rad/s)'] = dataframe['Speed (RPM)']*0.104719755
                x = dataframe[['Time (sec)', 'GV', 'I (A)', 'Speed (rad/s)', 'Flow Rate (GPM)', 'DP (psi)','torque (mNm)', 'V (V)' ]]
                self.dynamic_data.append(x)
                self.filenames.append(file_name)
                print(file_name)
                print(dataframe.shape)
        print()
            
        self.DP_model = DP_model
        self.torque_model = torque_model
        self.DP_model_GP = DP_model_GP
        self.torque_model_GP = torque_model_GP
        self.weight = weight
        
        self.predicted_data=[]
        self.start = start
        self.objectives = objectives
        self.progress = 0
        
    def get_prediction(self, model_parameters):
        
        self.predicted_data = []
        
        dynamic_model = turbine_dynamic_model(Tw = model_parameters[0] , H = model_parameters[1])
        
        steady_state_model = turbine_steady_state_model(self.DP_model, self.torque_model,\
                                                        self.DP_model_GP, self.torque_model_GP,\
                                                        torque_constant = model_parameters[2],\
                                                        torque_resistance = model_parameters[3],\
                                                        speed_constant = model_parameters[4],\
                                                        internal_resistance = model_parameters[5])
        
        for i, runs in enumerate(self.dynamic_data):
            #print("running file: " + self.filenames[i])
            data = runs.values
            steady_state_model.update_states(G=data[0:self.start,1].mean(), w=data[0:self.start,3].mean(),\
                                             q=data[0:self.start,4].mean(), I=data[0:self.start,2].mean())
            
            t= data[self.start:,0]
            G_t = data[self.start:,1]
            I_t = data[self.start:,2]
            
            dynamic_model.simulate(steady_state_model, t, G_t, I_t)
            df = pd.DataFrame(dynamic_model.state_t, columns =['Time (sec)', 'GV', 'Speed (rad/s)', 'Flow Rate (GPM)',\
                                                               'DP (psi)','torque mech (mNm)', 'I (A)', 'V (V)',\
                                                               'torque (mNm)' ]) 
            self.predicted_data.append(df)
            
    def prediction_error(self, weight = 1):
        error = 0
        for i, runs in enumerate(self.dynamic_data):
            #print("evaluating file: " + self.filenames[i])
            size = len(self.predicted_data[i][:]['Time (sec)'].values)
            for j, variable in enumerate(self.objectives):
                error += self.weight[j] * np.sum((runs[self.start:][variable].values - self.predicted_data[i][:][variable].values)**2) / size
        return [error]
    
    def fitness(self,model_parameters):
        param = []
        for p in model_parameters:
            param.append(p)
        try:
            self.get_prediction(param)
            error = self.prediction_error()
        except:
            print(param)
            error = [100000]
            
        self.progress += 1
        if self.progress%32 == 0:
            print(self.progress//32)
        return error
    
    def get_bounds(self):
        return ([1,0.1,40,1,200,0.5],[10,5,50,30,300,10]) #problem.evaluate_model_parameters([2, 0.8, 46.4, 6.3329, 206, 55.312])
```
