```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from smt.surrogate_models import KRG
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import (RBF, Matern, RationalQuadratic,
                                              ExpSineSquared, DotProduct,
                                              ConstantKernel)
```

```python
path=r"C:\Users\lilly\OneDrive\Documents\1.0_Graduate_Studies\5.0 Energy havester\5.8_code\Energy_Harvester\Data"
flow27 = pd.read_csv(path+r"\2019_01_13_a_SS.csv")
flow27['Speed (rad/s)'] = flow27['Speed (RPM)']*0.104719755
flow27
```

```python
path=r"C:\Users\lilly\OneDrive\Documents\1.0_Graduate_Studies\5.0 Energy havester\5.8_code\Energy_Harvester\Data"
flow26 = pd.read_csv(path+r"\2019_01_13_d_SS.csv")
flow26['Speed (rad/s)'] = flow26['Speed (RPM)']*0.104719755
flow26
```

```python
def train_kriging(data):
    
    trigger=False
    xt=None
    yt_DP=None
    yt_torque=None
    
    for sheet in data:
        x = sheet[['Speed (rad/s)','GV','Flow Rate (GPM)']]
        y1 = sheet[['DP (psi)']]
        y2 = sheet[['torque (mNm)']]
        
        if trigger:
            xt = np.concatenate((xt,np.array(np.array(x.values))),axis=0)
            yt_DP = np.concatenate((yt_DP,np.array(np.array(y1.values))),axis=0)
            yt_torque = np.concatenate((yt_torque,np.array(np.array(y2.values))),axis=0)
        else:
            trigger=True
            xt = np.array(np.array(x.values))
            yt_DP = np.array(np.array(y1.values))
            yt_torque = np.array(np.array(y2.values))
            
    sm_DP = KRG(theta0=[0.1,0.1,0.1],poly='quadratic',print_global=False)  
    sm_DP.set_training_values(xt, yt_DP)
    sm_DP.train()

    sm_torque = KRG(theta0=[0.1,0.1,0.1],poly='quadratic',print_global=False)  
    sm_torque.set_training_values(xt, yt_torque)
    sm_torque.train()
                      
    return [sm_DP,sm_torque,xt,yt_DP,yt_torque]
```

```python
[sm_DP,sm_torque,xt,yt_DP,yt_torque] = train_kriging([flow27,flow26])
print(sm_DP.predict_values(np.array([[416.9,15,26],[409,15,26]])))
print(sm_torque.predict_values(np.array([[416.9,15,26],[409,15,26]])))
```

```python
kernel= 1.0 * RBF(length_scale=1.0, length_scale_bounds=(1e-1, 10.0))
gp = GaussianProcessRegressor(kernel=kernel)
gp.fit(xt, yt_DP)
```

```python
y_mean, y_std = gp.predict(np.array([[416.9,5,27],[409,15,26]]), return_std=True)
```
